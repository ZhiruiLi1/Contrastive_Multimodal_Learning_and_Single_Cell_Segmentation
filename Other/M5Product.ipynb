```python
import json
from itertools import islice
```

```python
with open('product1m_product5m_train_id_label.json', 'r') as file:
    data = json.load(file)
```

```python
for key, value in islice(data.items(), 10):
    print(f"{key}: {value}")
    print("")
```

```python
with open('product1m_product5m_test_id_label.json', 'r') as file:
    data2 = json.load(file)
```

```python
for key, value in islice(data2.items(), 100):
    print(f"{key}: {value}")
    print("")
```

```python
for key, value in islice(data2.items(), 10):
    for key2, value2 in islice(value.items(), 10):
        print(f"{key2}: {value2}")
    print("")
```
